```python
import pandas as pd
import numpy as np
import seaborn as sns
black=pd.read_csv('BlackFriday.csv')
black.head(5)
```

```python
black.drop(['User_ID','Product_ID','Stay_In_Current_City_Years'],axis=1,inplace=True)
black.head(5)
```

```python
black['City_Category']=black['City_Category'].fillna("xxx")
black.head(5)
```

```python
black['City_Category']=black['City_Category'].map({
    'A':'Metro Cities',
    'B':'Small towns',
    'C':'Villages'
})
black.head(5)
```

```python
black.rename(columns={
    'Product_Category_1':'Baseball caps',
    'Product_Category_2':'Wine Tumblers',
    'Product_Category_3':'Pet Raincoats'
})
```

```python
black['Marital_Status']=black['Marital_Status'].map({
    0:'Unmarried',
    1:'Married'
})
black.head(5)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
ax=sns.countplot(x="Product_Category_1",hue='Gender',palette="Set1",data=black)
ax.set(title="Cat1",xlabel=" Bought Product category 1",ylabel="Total")
plt.show()
```
